```python
import threading
import queue
from queries import user, db
from importlib import reload
reload(db)

class SetQueue(queue.Queue):
    def _init(self, maxsize):
        self.queue = set()
    def _put(self, item):
        self.queue.add(item)
    def _get(self):
        return self.queue.pop()

round = 1
def worker(q, round):
    while True:
        item = q.get()
        print(f'Working on {item}')
        scrape_user(item, round)
        print(f'Finished {item}')
        q.task_done()
        
def supervisor(q, db,round):
    while True:
        if q.qsize() < 100:
            remaining = populateq(q, db, round)
            if remaining == 0:
                return
                
            
def populateq(q, db, round):
    db = db.Db()
    usrs = db.get_unscraped_nodes(round)
    db.close()
    for usr in usrs:
        q.put(usr)
    return len(usrs)

q = SetQueue() 
threading.Thread(target=worker, args=(q, round)).start()

threading.Thread(target=supervisor, args=(q, db, round)).start()

q.join()
```
